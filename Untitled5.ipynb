```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import  seaborn as sns
data = pd.read_csv('/content/StudentsPerformance.csv')
```

```python
columns = data.columns
print(columns)
```

```python
means = data.mean()
print(means)
```

```python
null_counts = data.isnull().sum()
print(null_counts)
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(data['math score'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Math Scores')
plt.xlabel('Math Score')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(y='parental level of education', data=data, palette='pastel')
plt.title('Count of Students by Parental Education Level')
plt.xlabel('Count')
plt.ylabel('Parental Education Level')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='gender', y='reading score', data=data, palette='pastel')
plt.title('Reading Score Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('Reading Score')
plt.show()
```
